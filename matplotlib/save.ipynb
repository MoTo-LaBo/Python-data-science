# plot を保存する
- サイズを指定して　plot する
- png ファイルで保存
- jpg ファイルで保存

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-3, 3, 10)
# dpi = dot per inch / 1inch あたりの dot size
plt.figure(figsize=(10, 5), dpi=50)
plt.plot(x, x**2)
```

```python
# object指向で記述する場合
fig, axes = plt.subplots(figsize=(10, 5))
axes.plot(x, x**2)
```

## png file での保存　:　.savefig( )

```python
df = pd.read_csv('../pandas/dataset/tmdb_5000_movies.csv')
plt.scatter('runtime', 'budget', data=df, alpha=0.2)
plt.xlabel('runtime')
plt.ylabel('budget')
plt.savefig('runtime_budget_scatter.png')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 5))
x = np.linspace(-3, 3, 100)
axes[0].plot(x, np.cos(x))
axes[0].set_title('x_cos')
axes[1].plot(x, np.sin(x))
axes[1].set_title('y_sin')
fig.tight_layout()
fig.savefig('cos_sin.png')
```

## PDF で保存

```python
from matplotlib.backends.backend_pdf import PdfPages

pdf = PdfPages('pdf_files.pdf')
fig, axes = plt.subplots()
axes.plot(x, x**2)
pdf.savefig(fig)
pdf.close()
```

```python
pdf = PdfPages('savemultifig_sample.pdf')
for i in range(10):
    fig, axes = plt.subplots()
    axes.plot(x, x**i)
    axes.set_title('x^{}'.format(i))
    pdf.savefig(fig)
pdf.close()
```
