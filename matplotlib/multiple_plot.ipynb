# 複数のグラフ（図）をplot する

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## plt.subplot( )
- グラフを横並びにする
- (　行, 列, index )

```python
x = np.linspace(-10, 10, 100)
```

```python
plt.plot(x, x**2)
```

```python
plt.plot(x, x**3)
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, x**2)
plt.subplot(1, 2, 2)
plt.plot(x, x**3)
```
