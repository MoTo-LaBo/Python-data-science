# DataFrame の filter

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'name': ['John', 'Zack', 'Emily'],
    'sex': ['male', 'male', 'female'],
    'age': [22, 30, 32]
}
df = pd.DataFrame(data)
df
```

```python
df[[True, False, True]] # list型で　True or Fales が　filter
```

```python
filter_series = pd.Series([True, False, True])
filter_series
```

```python
df[filter_series]
```

- 元の　df(DataFrame)の　index　がそのまま引き継がれている
    - *True　は表示されて　False　は表示されない*

```python
df['age'] >= 30
```

```python
filter_s = df['age'] >= 30
```

```python
df[filter_s]
```

```python
# 普通はこういう記述の仕方をするのが一般的
df[df['age'] >= 30]
```

#### 複数の条件をつける時
- **(　)&(　)　や　(　)|(　)　で複数の条件を入れる**
    - *(　)&(　)*　:　A　と　B　どちらも
    - *(　)|(　)*　:　A　または　B　どちらか

```python
# 予算(budget)が　0　もしくは、売り上げ(revenue)が　０のモノ
df = df[(df['budget'] == 0) | (df['revenue'] == 0)]
df[['budget', 'revenue']]
```

## ~ で　NOT　演算

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
# 予算も売り上げも 0 ではない
df = df[~((df['budget'] == 0) | (df['revenue'] == 0))]
df[['budget', 'revenue']]
```

- **df[ ~ ((df[ 'budget' ] == 0) | (df[ 'revevnue' ] == 0)) ]**
    - *~ (　)*　で　df[(df['budget'] == 0) | (df['revenue'] == 0)]　を囲むことによって、条件のNOT演算になる

```python
data = [
    {'id': 'a', 'bool': True},
    {'id': 'b', 'bool': False},
    {'id': 'c', 'bool': True}
]
```

```python
df = pd.DataFrame(data)
df
```

```python
df['bool']
```

```python
df[df['bool']]
```

```python
# False だけが必要な場合
df[~df['bool']]
```

## dropna( )
- *NaN を含んでいるレコード全て　drop*

```python
df.dropna()
```

```python
df.describe()
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
df.head(2)
```

```python
# runtime が　NaN であったら True になるような filter を作成
df[np.isnan(df['runtime'])]  # np.isnan( )で使用できるのは数値項目のみ
```

```python
# subset引数で特定のカラムに　NaN があるレコードだけ drop 可能
len(df.dropna(subset=['runtime']))
```

```python
len(df)
```

- 元の DataFrame は更新されないので注意

## .fillna( )

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
df['runtime']
```

```python
df['runtime'].mean()
```

- *この値を　runtime　が　NaN　のところに入れていく*

```python
df['runtime'] = df['runtime'].fillna(df['runtime'].mean())
```

```python
df[np.isnan(df['runtime'])]
```
