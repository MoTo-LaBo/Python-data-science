# DataFrame　と　Series　の便利関数

### .uique( ) と　.nunique( )
- Series に対して使用できるモノ

```python
import numpy as np
import pandas as pd
```

```python
df_movies = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
df_movies['original_language'].unique()
```

- original language のユニークな値のみを取得

- *ユニーク（英：unique）とは*

    - **すべての値がオンリーワン/重複しない**

```python
df_movies['original_language'].nunique()
```

- nunique( )
    - **ユニークな値の数を取得**

```python
df_movies['id'].nunique()
```

```python
len(df_movies)
```

## .value_counts( )
- Series に対して使用可能
- **それぞれの値に幾つのレコードがあるかを取得**

```python
df_movies['original_language'].value_counts()
```

## .sort_values()
- DataFrame に使用可能
- sort : 並び替え


- <u>指定した columns で sort してくれる</u>
    - **default は昇順(ascending)**

```python
df_movies.sort_values('budget')
```

```python
df_movies.sort_values('budget', ascending=False) # 降順になる
```
