# DataFrame の iteration

## .apply
- 関数を各業に適用(apply)する

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
# revenue(興行収入) - budget(予算)　= 利益
df.head(1)
```

```python
# return_nan_if_zero(0) --> nan
# return_nan_if_zero(10000) --> 10000
def return_nan_if_zero(budget):
    if budget == 0:
        return np.nan
    else:
        return budget

    
return_nan_if_zero(0)
```

```python
df['budget']
```

```python
df['budget'].apply(return_nan_if_zero)
```

- .apply( )
    - *budget columns の　Series　に　apply　すれば*
        - **各　budget　に対して関数を実行した値が　Series　で返ってくる**

```python
df['budget'] = df['budget'].apply(return_nan_if_zero)
```

```python
df
```

- **apply　した　budget　columns　を元の　DataFrame　の　budget columns　に上書き**
    - *しっかりと NaN になっている*

```python
budget = 0
```

```python
np.nan if budget == 0 else budget
```

```python
def return_nan_if_zero(budget):
    
    return np.nan if budget == 0 else budget
```

```python
return_nan_if_zero(0)
```

```python
return_nan_if_zero(100)
```

### 上記を　lambda(ラムダ)関数にする

```python
x = lambda budget: np.nan if budget == 0 else budget
```

```python
x(budget)
```

```python
budget = 100
```

```python
x(budget)
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
df['budget'] = df['budget'].apply(lambda budget: np.nan if budget == 0 else budget)
df['revenue'] = df['revenue'].apply(lambda revenue: np.nan if revenue == 0 else revenue)
```

```python
df
```

- *簡単な関数であれば、関数を作成するのではなく*
    - **lambda 関数を使用して　１行で記述する**

```python
df[['budget', 'revenue']].dropna()
```

```python
df = df.dropna(subset=['budget', 'revenue'])
```

```python
df
```

### revenue(興行収入) - budget(予算)　= 利益

```python
df.apply(lambda row: row['revenue'] - row['budget'], axis=1)  # lambda に row を指定する時は axis=1 とする
```

```python
df['plofit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)
```

```python
df['plofit'].max()
```

```python
df['plofit'].idxmax()
```

```python
df.iloc[df['plofit'].idxmax()]
```

```python
df[df['plofit'] == df['plofit'].max()] # DataFrame を取得　/ filter を使用
```

## .iterrows( )

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
next(df.iterrows())
```

```python
a
```

```python
b
```

- index と　各行が返される


- **.iterrows( )**
    - *row をイテレーションさせる。　イテレーターは　( idx, row )　を返す*

```python
for _, row in df.iterrows():
    # row の vote_averageが　１０だったら、タイトル表示
    if row['vote_average'] == 10:
        print(row['title'])
```

- *今回　idx　を使用していないので　_(アンダースコア)　とする*
    - **idx などの変数を for文内で利用しない場合は　_　(アンダースコア)　とする**
