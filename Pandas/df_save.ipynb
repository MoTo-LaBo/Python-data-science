# DataFrame の保存

## .to_csv( )

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
df.head(3)
```

- profit(revenue - budget) の高い順い並べた表を作成してみる (budget や revenueが　０のものは除く)

```python
df['budget'] = df['budget'].apply(lambda x: np.nan if x == 0 else x)
df['revenue'] = df['revenue'].apply(lambda x: np.nan if x == 0 else x)
df = df.dropna(subset=['budget', 'revenue'])
```

```python
df['plofit'] = df.apply(lambda x: x['revenue'] - x['budget'], axis=1)
```

```python
df = df['plofit'].sort_values(ascending=False) 
```

## 回答

```python
# 1. csv file　読み込み
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
df.head(2)
```

```python
# 2. budget や revenue が 0 のモノを filter out
df = df[~((df['budget'] == 0) | (df['revenue'] == 0))]  # NOT 演算
```

```python
# 3. plofit columns を作成
df['plofit'] = df.apply(lambda x: x['revenue'] - x['budget'], axis=1)
```

```python
# 4. plofit で　sort　する
df = df.sort_values('plofit', ascending=False)
```

```python
df.head(3)
```

```python
df.to_csv('dataset/tmdb_5000_movies_plofit_sorted.csv', index=False)
```

- **.to_csv( 'path' )**
    - *DataFrame を csv 形式で保存*
    

- **index=False**
    - 保存後　→ 余計な　index をつけいないようにする為
        - read すると新たに　index　が付与されるので

```python
updated_df = pd.read_csv('dataset/tmdb_5000_movies_plofit_sorted.csv')
```
